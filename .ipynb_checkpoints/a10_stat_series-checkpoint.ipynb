```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
from scipy.stats import boxcox
import statsmodels.api as sm
import statsmodels.tsa.api as smt

path_base = 'data/'

sales_of_company_x = pd.read_csv(path_base + "monthly-sales-of-company-x-jan-6.csv")
robberies_in_boston = pd.read_csv(path_base + "monthly-boston-armed-robberies-j.csv")
airlines_passengers = pd.read_csv(path_base + "international-airline-passengers.csv")
mean_monthly_temp = pd.read_csv(path_base + "mean-monthly-air-temperature-deg.csv")
dowjones_closing = pd.read_csv(path_base + "weekly-closings-of-the-dowjones-.csv")
female_births = pd.read_csv(path_base + "daily-total-female-births-in-cal.csv")

import matplotlib.pyplot as plt
%matplotlib inline

all_series = {
    "Monthly sales of company X": sales_of_company_x["Count"],
    "Monthly Boston armed robberies": robberies_in_boston["Count"],
    "International airline passengers: monthly totals in thousands": airlines_passengers["Count"],
    "Mean monthly air temperature (Deg. F) Nottingham Castle": mean_monthly_temp["Deg"],
    "Weekly closings of the Dow-Jones industrial average": dowjones_closing["Close"],
    "Daily total female births in California": female_births["Count"]
}

stationary_series = {}

series_p_values = {}

# with plt.style.context('bmh'):
#     plt.figure(figsize=(16, 8))
#     layout = (3, 2)
#     for i, key in enumerate(all_series.keys()):
#         x = i % 2
#         y = int((i - x) / 2)
        
#         ts_ax = plt.subplot2grid(layout, (y, x))
#         all_series[key].plot(ax=ts_ax, color='blue')
#         ts_ax.set_title(key)
        
#     plt.tight_layout()

def test_stationarity(timeseries):
    print('Results of Dickey-Fuller Test:')
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic', 'p-value', '#Lags Used', 'Number of Observations Used'])
    for [key, value] in dftest[4].items():
        dfoutput['Critical Value (%s)' % key] = value
    print(dfoutput)

def tsplot(y, lags=None, figsize=(14, 8), style='bmh'):
    # test_stationarity(y)
    if not isinstance(y, pd.Series):
        y = pd.Series(y)
    with plt.style.context(style):
        plt.figure(figsize=figsize)
        layout = (4, 1)
        ts_ax = plt.subplot2grid(layout, (0, 0), rowspan=2)
        acf_ax = plt.subplot2grid(layout, (2, 0))
        pacf_ax = plt.subplot2grid(layout, (3, 0))

        y.plot(ax=ts_ax, color='blue', label='Or')
        ts_ax.set_title('Original')

        smt.graphics.plot_acf(y, lags=lags, ax=acf_ax, alpha=0.05)
        smt.graphics.plot_pacf(y, lags=lags, ax=pacf_ax, alpha=0.05)

        plt.tight_layout()
    return
    
def adf_test(series, print_required):
    """
       -
    :param series:    
    :return:  
    """
    result = adfuller(series, autolag='AIC')

    if print_required:
        print('ADF : {:.4f}'.format(result[0]))
        print('p-: {:.4f}'.format(result[1]))
        print(' :')
        for key, value in result[4].items():
            print('\t{}: {:.4f}'.format(key, value))
            
        # 
        if result[1] <= 0.05:
            print(":    (  )")
        else:
            print(":    (    )")

    return float(result[1])

#    Monthly sales of company X
series = boxcox(all_series["Monthly sales of company X"], 0)
series = series[1:] - series[:-1]
stationary_series["Monthly sales of company X"] = pd.Series(series)
series_p_values["sales p-value"] = adf_test(stationary_series["Monthly sales of company X"], 0)

#    Monthly Boston armed robberies
series = boxcox(all_series["Monthly Boston armed robberies"], 0)
series = series[1:] - series[:-1]
stationary_series ["Monthly Boston armed robberies"] = pd.Series(series)
series_p_values["robberies p-value"] = adf_test(stationary_series["Monthly Boston armed robberies"], 0)

#    International airline passengers: monthly totals in thousands
series = boxcox(all_series["International airline passengers: monthly totals in thousands"], 0)
series = series[1:] - series[:-1]
series = series[12:] - series[:-12]
stationary_series["International airline passengers: monthly totals in thousands"] = pd.Series(series)
series_p_values["passengers p-value"] = adf_test(stationary_series["International airline passengers: monthly totals in thousands"], 0)

#      p-value      , ..    
stationary_series["Mean monthly air temperature (Deg. F) Nottingham Castle"] = all_series["Mean monthly air temperature (Deg. F) Nottingham Castle"]
series_p_values["Temperature p-value"] = adf_test(stationary_series["Mean monthly air temperature (Deg. F) Nottingham Castle"], 0)

#    Weekly closings of the Dow-Jones industrial average
series = boxcox(all_series["Weekly closings of the Dow-Jones industrial average"], 0)
series = series[1:] - series[:-1]
stationary_series["Weekly closings of the Dow-Jones industrial average"] = pd.Series(series)
series_p_values["Dow-Jones p-value"] = adf_test(stationary_series["Weekly closings of the Dow-Jones industrial average"], 0)

#      p-value      , ..    
stationary_series["Daily total female births in California"] = all_series["Daily total female births in California"]
series_p_values["Female births p-value"] = adf_test(stationary_series["Daily total female births in California"], 0)

#  DataFrame      
df = pd.DataFrame({
    'Series': list(series_p_values.keys()),
    'P-value': list(series_p_values.values())
})

print(df)

#  
n_rows = len(all_series)
fig, axes = plt.subplots(n_rows, 2, figsize=(16, 4*n_rows))

#    , ,  axes -  
if n_rows == 1:
    axes = axes.reshape(1, 2)

#    
for i, key in enumerate(all_series.keys()):
    #  :  
    if hasattr(all_series[key], 'plot'):
        all_series[key].plot(ax=axes[i, 0], color='blue')
    else:
        axes[i, 0].plot(all_series[key], color='blue')
    
    axes[i, 0].set_title(f": {key}")
    axes[i, 0].grid(True)
    
    #  :  
    if hasattr(stationary_series[key], 'plot'):
        stationary_series[key].plot(ax=axes[i, 1], color='green')
    else:
        axes[i, 1].plot(stationary_series[key], color='green')
    
    axes[i, 1].set_title(f": {key}")
    axes[i, 1].grid(True)

#    
plt.tight_layout()
plt.show()
```
